```python
import nltk
import pandas as pd
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import stopwords
import re
```

```python
data=pd.read_csv("Restaurant_Reviews.tsv")
```

```python
data.info()
```

```python
data.head()
```

```python
data["Review"][0]
```

```python
data["review_lowercase"]=data["Review"].str.lower()
```

```python
en_stopwords=stopwords.words("english")
```

```python
en_stopwords.remove("not")
```

```python
print(en_stopwords)
```

```python
data["no_stopwords"]=data["review_lowercase"].apply(lambda x: ' '.join([word for word in x.split() if word not in (en_stopwords)]))
```

```python
print(data["no_stopwords"][0])
```

```python
data["no_stopwords_no_puc"]=data.apply(lambda x: re.sub(r"[*]","star",x["no_stopwords"]),axis=1)
```

```python
data.head()
```

```python
data["no_stopwords_no_puc"]=data.apply(lambda x: re.sub(r"([^\w\s])","",x["no_stopwords_no_puc"]),axis=1)
```

```python
data["tokenize"]=data.apply(lambda x: word_tokenize(x["no_stopwords_no_puc"]),axis=1)
```

```python
data.head()
```

```python
data["tokenize"][0]
```

```python
ps=PorterStemmer()
```

```python
data["stemmed"]=data["tokenize"].apply(lambda tokens: [ps.stem(token)for token in tokens])
```

```python
data.head()
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
data["lemmatized"]=data["tokenize"].apply(lambda tokens: [lemmatizer.lemmatize(token) for token in tokens])
```

```python
tokens_clean=sum(data["lemmatized"],[])
```

```python
unigrams=(pd.Series(nltk.ngrams(tokens_clean,1)).value_counts())
```

```python
print(unigrams[:10])
```

```python
bigrams=(pd.Series(nltk.ngrams(tokens_clean,2)).value_counts())
```
